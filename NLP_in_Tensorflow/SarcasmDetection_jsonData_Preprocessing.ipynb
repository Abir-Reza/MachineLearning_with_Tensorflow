#A Dataset containing(**json data**) **News-hedline, url-link and label** that is sarcasm or not. 

Preprocessing data from the dataSet.

```python
# Download dataset
!wget --no-check-certificate \
    https://storage.googleapis.com/laurencemoroney-blog.appspot.com/sarcasm.json \
    -O /tmp/sarcasm.json
  

#This allows to load data in JSON format and automatically create a Python data structure from it
import json

with open("/tmp/sarcasm.json",'r') as f:
  datastore =json.load(f)


sentences = []
labels = []
urls = []

# datastore[2]

for item in datastore :
  sentences.append(item['headline'])
  labels.append(item['is_sarcastic'])
  urls.append(item['article_link'])
```

#Preprocessing

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(oov_token = '<OOV>')
tokenizer.fit_on_texts(sentences)

word_index =tokenizer.texts_to_sequences(sentences)
print(word_index)
print('The length of the word_index(Total_number_of_sentences) is : ',len(word_index))


sequences = tokenizer.texts_to_sequences(sentences)
#padding 0 at the last of the sentences, 
#and maximum length of a sentence is 30
padded = pad_sequences(sequences,padding='post', maxlen=30)

print(padded[0])
print(padded.shape)
```
